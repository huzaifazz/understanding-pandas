# How to rename columns?

```python
df = pd.DataFrame({'X': [0, 0, 0, 0, 1, 1, 1, 1,], 'Y': [0, 0, 1, 1, 0, 0, 1, 1], 'Z': [0, 1, 0, 1, 0, 1, 0, 1]})
df
```

```python
df.rename(columns={'X': 'X_col', 'Y': 'Y_col', 'Z': 'Z_col'}, inplace=True)
df
```

```python
# Other remaning method
df.columns={'X-col','Y-col','Z-col'}
df
```

```python
# to change some char or string
df.columns=df.columns.str.replace('-', '--')
df
```

```python
# Adding suffix or prefix
df = df.add_prefix('>')
df
```
